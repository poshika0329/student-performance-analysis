# Student Performance Analysis using Python

## Problem Statement
The objective of this project is to analyze student performance data and identify patterns, relationships, and key factors affecting scores using data analysis techniques.

## Steps Performed
- Data loading and preprocessing  
- Exploratory Data Analysis (EDA)  
- Data visualization  
- Insight generation  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset (same but slightly extended)
data = {
    'gender': ['female','male','female','male','female','male','female','male','female','male'],
    'math score': [72,69,90,47,76,71,88,40,65,85],
    'reading score': [72,90,95,57,78,83,95,43,70,88],
    'writing score': [74,88,93,44,75,78,92,39,68,85]
}

df = pd.DataFrame(data)

# Summary
print(df.describe())
```

```python
sns.histplot(df['math score'], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
sns.boxplot(data=df[['math score','reading score','writing score']])
plt.title("Score Comparison")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.barplot(x='gender', y='math score', data=df)
plt.title("Math Score by Gender")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

INSIGHTS FROM EXPLORATORY DATA ANALYSIS

A strong positive correlation exists between reading and writing scores, indicating that students who perform well in reading also tend to perform well in writing.

Math scores exhibit greater variability compared to reading and writing, suggesting inconsistent performance among students in mathematics.

Gender-based analysis shows that female students tend to achieve higher scores in reading and writing, while math performance remains relatively balanced across genders.

The dataset does not contain significant outliers, indicating a well-distributed and reliable dataset for analysis.

Overall, all three subjects demonstrate positive relationships, implying that improvement in one subject is associated with better performance in others.


## Conclusion
The analysis reveals strong relationships between reading and writing scores and moderate variation in math performance. Gender-based trends indicate differences in subject-wise performance. These insights can help in understanding student strengths and improving academic strategies.
